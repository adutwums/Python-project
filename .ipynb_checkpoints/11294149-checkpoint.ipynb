```python
class redWine:
    def __init__(self, data):
        self.data = pd.read_csv(data)
    
    def scatter_plot(self):
        plt.scatter(self.data['pH'],self.data['residual sugar'],color= 'blue',alpha=0.7)
        plt.title('pH vs. residual sugar')
        plt.xlabel('pH')
        plt.ylabel('residual sugar')
        plt.grid(True)
        plt.show()

    def histogram(self):
        plt.figure(figsize=(8, 8))
        plt.hist(self.data['residual sugar'], bins=20, color='blue', edgecolor='black')
        plt.title('The residual sugar content')
        plt.xlabel('residual sugar Content')
        plt.ylabel('Frequency')
        plt.show()
   
    def correlation(self):
        correlation  = self.data.corr()
        plt.figure(figsize=(8, 8))
        plt.imshow(correlation, cmap='coolwarm', interpolation='nearest')
        plt.colorbar()
        plt.title('Correlation Plot')
        plt.xticks(range(len(correlation)), correlation.columns, rotation=90)
        plt.yticks(range(len(correlation)), correlation.columns)
        plt.tight_layout()
        plt.show()

dataset = redWine('red_wine.csv')
```

## Scatter Plot Diagram

```python
dataset.scatter_plot()
```

##### Conclusion : The scatter plot shows the relationship between pH and residual sugar. Residual sugar is high at a pH range between 3.0 and 3.4 and lowest between 2.8 and 3.0.

## Bar Chart Diagram

```python
dataset.histogram()
```

#### Conclusion : The diagram above is a bar chart showing the residual content of wine against the frequency.

## Correlation Plot

```python
dataset.correlation()
```

#### Conclusion : The diagarm above is a correlation table of the data. The data is a negative correlation.
